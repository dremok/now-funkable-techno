```python
import math
import cv2
from matplotlib import pyplot as plt
```

```python
HEIGHT = 256
WIDTH = 256
```

```python
img = cv2.imread('data/hearthstone/AT_023.png', cv2.IMREAD_UNCHANGED)
img = cv2.resize(img, (169, 256), interpolation = cv2.INTER_AREA)
plt.imshow(img)
```

```python
img_padded = cv2.copyMakeBorder(img, 0, HEIGHT - img.shape[0], math.floor((WIDTH - img.shape[1]) / 2), math.ceil((WIDTH - img.shape[1]) / 2), cv2.BORDER_REPLICATE)
```

```python
import os
```

```python
DATA_DIR = 'data/hearthstone'
PADDED_DIR = f'data/hs_256'

os.makedirs(PADDED_DIR, exist_ok=True)

for filename in os.listdir(DATA_DIR):
    if not filename.endswith('.png'):
        continue
    img = cv2.imread(f'{DATA_DIR}/{filename}', cv2.IMREAD_COLOR)
    img = cv2.resize(img, (169, 256), interpolation = cv2.INTER_AREA)
    img_padded = cv2.copyMakeBorder(img, 0, HEIGHT - img.shape[0], math.floor((WIDTH - img.shape[1]) / 2), math.ceil((WIDTH - img.shape[1]) / 2), cv2.BORDER_REPLICATE)
    res = cv2.imwrite(f'{PADDED_DIR}/{filename}', img_padded)
    if res:
        pass
        print(f'Padded {filename} successfully saved!')
    else:
        print(res)
```
